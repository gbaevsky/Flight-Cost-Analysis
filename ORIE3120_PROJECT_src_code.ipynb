```python
import seaborn as sb

import statsmodels.api as sm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
```

```python
fly = pd.read_csv("Clean_Dataset.csv")
fly.head()
```

```python
#get median price as threshold
price = fly["price"]
threshold = price.median()
print(threshold)
```

```python
# replace economy = 1 and business = 0
fly["price"] = (price>threshold).replace({True:1, False:0})
fly["price"]
```

```python
fly["stops"].unique()
```

```python
fly["stops"] = fly["stops"].replace({"zero":0, "one":1, "two_or_more":2})
```

```python
a = [ fly[fly["airline"] ==airline] for airline in fly["airline"].unique()]
print(a)
```

```python
a1 = a[0]
a2 = a[1]
a3 = a[2]
a4 = a[4]
a5 = a[5]

f = plt.figure()
f.set_figwidth(12)
  
sb.regplot(x=a1["stops"], y=a1["price"], data=a1, logistic=True, ci=None,
             scatter_kws={'color': 'black'})
sb.regplot(x=a2["stops"], y=a2["price"], data=a1, logistic=True, ci=None,
             scatter_kws={'color': 'black'})
sb.regplot(x=a3["stops"], y=a3["price"], data=a1, logistic=True, ci=None,
             scatter_kws={'color': 'black'})
sb.regplot(x=a4["stops"], y=a4["price"], data=a1, logistic=True, ci=None,
             scatter_kws={'color': 'black'})
sb.regplot(x=a5["stops"], y=a5["price"], data=a1, logistic=True, ci=None,
             scatter_kws={'color': 'black'})

plt.legend(["SpiceJet","AirAsia","Vistara","GO_FIRST","Indigo","Air_India"  ])
plt.xticks([0,1,2])
plt.yticks([0,1])
plt.ylabel("price > median (7425.0)")
plt.title("Logistic Regression of Price vs. Number of Stops")
```
